Se necesita el listado de contactos de aquellos usuarios mayores de 30 años que se unieron este
año, ordenados por fecha de registro, del más antiguo al más nuevo. Solo es necesario mostrar los
primeros 5 registros encontrados. En este listado debe aparecer mínimamente el nombre, el
correo, el número celular, la dirección y la fecha de registro.

```python
query = """
SELECT users.name, users.email, profiles.phone, profiles.address, users.created_at
FROM users
INNER JOIN profiles ON users.user_id = profiles.user_id
WHERE users.created_at > '2025-01-01' AND users.age > 30
ORDER BY users.created_at ASC
LIMIT 5
"""

pd.read_sql_query(query, conn)
```

Entregue un resumen de cuánto ha gastado cada cliente en total, pero solo de quienes hayan
superado los 700 dólares en su historial de compras.

```python
query = """
SELECT users.name, SUM(orders.amount) AS total_spent
FROM users
INNER JOIN orders ON users.user_id = orders.user_id
GROUP BY users.user_id
HAVING total_spent > 1000
"""

pd.read_sql_query(query, conn)
```

Sería de interés para el negocio saber en qué ciudades potentes se tiene presencia: genere la lista
de ubicaciones únicas donde viven usuarios que hayan realizado al menos dos pedidos.

```python
query = """
SELECT profiles.address
FROM profiles
INNER JOIN orders ON profiles.user_id = orders.user_id
GROUP BY profiles.user_id
HAVING COUNT(orders.order_id) >= 2
"""

pd.read_sql_query(query, conn)
```

Se requiere de un desglose del número de usuarios en función de su grupo de edad: jóvenes,
adultos y seniors, según los rangos: menores de 25, entre 25 y 40, mayores de 40.

```python
query = """
SELECT 
    CASE 
        WHEN age < 25 THEN 'Jóvenes'
        WHEN age BETWEEN 25 AND 40 THEN 'Adultos'
        ELSE 'Seniors'
    END AS grupo_edad,
    COUNT(*) AS numero_usuarios
    FROM users
    GROUP BY grupo_edad
"""

pd.read_sql_query(query, conn)
```

Genere un reporte que incluya los datos de contacto y toda la información personal de cada
usuario, junto con sus pedidos más recientes. Muéstrelos ordenados cronológicamente.

```python
query = """
SELECT users.name, users.email, profiles.phone, profiles.address, orders.order_id, orders.product, orders.amount
FROM users
INNER JOIN profiles ON users.user_id = profiles.user_id
INNER JOIN orders ON users.user_id = orders.user_id
ORDER BY orders.order_date DESC
"""

pd.read_sql_query(query, conn)
```

```python
vn.train(
    ddl="""
CREATE TABLE users (
    user_id    INTEGER PRIMARY KEY,
    name       TEXT    NOT NULL,
    age        INTEGER,
    email      TEXT,
    created_at DATE
)
"""
)
```

```python
# https://vanna.ai/account/profile
api_key = '' # api key que entrega vanna para el uso de LLM
vanna_model_name = 'star_schema' # el nombre de la base de datos
vn1 = VannaDefault(model=vanna_model_name, api_key=api_key)
```

```python
vn1.train(
    ddl="""
    CREATE TABLE customer (
    customer_id INTEGER PRIMARY KEY,
    customer_name VARCHAR(100),
    customer_city VARCHAR(100)
    );
"""
)
```

```python
vn1.train(ddl="""
          CREATE TABLE product (
            product_id INTEGER PRIMARY KEY,
            product_name VARCHAR(100),
            category VARCHAR(100),
            quantity INTEGER,
            unit_price FLOAT
          );
""")
```

```python
vn1.train(ddl="""
          CREATE TABLE rep (
            rep_id INTEGER PRIMARY KEY,
            rep_name VARCHAR(100),
            region VARCHAR(100)
          );
""")
```

```python
vn1.train(ddl="""
            CREATE TABLE sales (
            sale_id INTEGER PRIMARY KEY,
            sale_date DATE,
            customer_id INTEGER,
            product_id INTEGER,
            rep_id INTEGER,
            quantity_sold INTEGER,
            total_amount FLOAT,
            FOREIGN KEY (customer_id) REFERENCES customer(customer_id),
            FOREIGN KEY (product_id) REFERENCES product(product_id),
            FOREIGN KEY (rep_id) REFERENCES rep(rep_id)
            );            
""")
```

```python
vn1.get_training_data()
```

```python
vn1.train(documentation="""
customer_id es el identificador único de cada cliente.
customer_name es el nombre del cliente.
customer_city es la ciudad del cliente.

product_id es el identificador único de cada producto.
product_name es el nombre del producto.
category indica la categoría del producto (por ejemplo, tecnología, hogar, etc).
quantity es la cantidad en inventario de ese producto.
unit_price es el precio unitario de cada producto.

rep_id es el identificador del representante de ventas.
rep_name es el nombre del representante.
region es la región en la que trabaja el representante.

sale_id es el identificador de cada venta realizada.
sale_date indica la fecha en que ocurrió la venta.
customer_id en la tabla sales indica qué cliente hizo la compra (relación con tabla customer).
product_id en la tabla sales indica qué producto se vendió (relación con tabla product).
rep_id en la tabla sales indica qué representante hizo la venta (relación con tabla rep).
quantity_sold es la cantidad de unidades vendidas en esa venta.
total_amount hace referencia al total de ventas para esa transacción (quantity_sold * unit_price).
""")
```

```python
query = vn1.generate_sql(
    "¿Cuál fue el total de ventas por cada ciudad de los clientes?"
)
```

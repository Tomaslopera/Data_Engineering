# Transformaci√≥n

## Tabla Codes

```python
df_codes_clean = df_codes.copy()
df_codes_clean = df_codes_clean[df_codes_clean['Code_comm'].notnull()] #¬†eliminar filas con c√≥digo nulo
df_codes_clean
```

```python
## Crear una funci√≥n que permita extraer los c√≥digos de productos
def clean_code(text):
    text = str(text)
    parent_code = None
    if len(text) == 11:
        code = text[:5]
        parent_code = text[:1]
    else:
        code = text[:6]
        parent_code = text[:2]
    try:
        parent = df_parents[df_parents['Code_comm'] == parent_code]['Description'].values[0]
    except:
        parent = None
    return (code, parent)
```

```python
df_codes_clean[['clean_code', 'parent_description']] = df_codes_clean.apply(lambda x: clean_code(x['Code']), axis=1, result_type='expand')
df_codes_clean
```

```python
df_codes_clean = df_codes_clean[df_codes_clean['clean_code'].notnull()][['clean_code', 'Description', 'parent_description']]
df_codes_clean
```

```python
dim_codes = df_codes_clean.copy()
dim_codes['id_code'] = df_codes_clean.index
dim_codes['clean_code'] = dim_codes['clean_code'].astype('int64')
dim_codes
```

### Creaci√≥n de dimensiones

```python
# Construir una funci√≥n que construya las dem√°s dimensiones
def create_dimensiones(data, id_name):
    list_keys = []
    value = 1
    for key in data:
        list_keys.append(value)
        value = value + 1
    
    return pd.DataFrame({id_name:list_keys, 'values':data})
```

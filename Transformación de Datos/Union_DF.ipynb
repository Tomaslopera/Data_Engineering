```python
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
transactions = pd.read_csv("../Transformación de Datos/Datasets/transactions.csv")
customers = pd.read_csv("../Transformación de Datos/Datasets/customers.csv")
products = pd.read_csv("../Transformación de Datos/Datasets/products.csv")
```

## Merge

```python
df_merge = transactions.merge(
    customers, 
    on="customer_id", # Merge with customers on customer_id
    how="inner").merge(
        products, 
        on="product_id", # Merge with products on product_id
        how="inner") # inner - left - right - outer

df_merge[["transaction_id","customer_id","first_name","product_name","amount"]]
```

## Join

```python
transactions_mod = transactions[["transaction_id","customer_id","product_id","amount"]].set_index("customer_id")
customer_mod = customers.set_index("customer_id")[["first_name","last_name"]]

df_join = transactions_mod.join(customer_mod, on="customer_id")
df_join
```

## Concat

```python
part1 = transactions.head(5)
part2 = transactions.tail(5)

df_concat_rows = pd.concat([part1, part2], axis=0)
df_concat_rows
```

```python
c1 = customers[["customer_id", "first_name"]].set_index("customer_id")
c2 = customers[["customer_id", "city"]].set_index("customer_id")

df_concat_columns = pd.concat([c1, c2], axis=1)
df_concat_columns
```
